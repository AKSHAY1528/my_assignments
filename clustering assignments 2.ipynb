# ****************Hirearichal Clustering********** 

```python
#libraries
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('crime_data.csv')
```

```python
df
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

#create dendogram


```python
dendro_gram=sch.dendrogram(sch.linkage(x,method='average'))
```

```python
#create no of clusters
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean')
```

```python
hc.fit_predict(x)
```

```python
y_hc = hc.fit_predict(x)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
df['h_clusterid'] = Clusters
```

```python
df.sort_values("h_clusterid")
```

```python
df.h_clusterid.value_counts()
```

# *******************************KMEANS*************************

```python
df
```

```python
X_km=df.drop('h_clusterid',axis=1)
```

```python
X_km
```

```python
XY=sc.fit_transform(X_km)
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=5,random_state=8)
model.fit(XY)
```

```python
wcss=[]
for i in range(1,12):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(XY)
    dist = kmeans.inertia_
    wcss.append(dist)
```

```python
plt.plot(range(1,12),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("Within a cluster sum of square")
plt.grid()
```

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=0)
clusters_new.fit(XY)
```

```python
X_km['new_clustres']=clusters_new.labels_
```

```python
X_km
```

```python
X_km.groupby('new_clustres').agg(['mean']).reset_index()
```
